##### 1. Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

##### 2. Importing the dataset

```python
dataset = pd.read_csv('1.01. Simple linear regression.csv')
# x = dataset.iloc[:, 0].values       # Independent variable - SAT Score
# y = dataset.iloc[:, 1].values       # Dependent variable - GPA Prediction

x = dataset[['SAT']]
y = dataset['GPA']

#dataset.head()
#dataset.shape
#dataset.count()
#dataset.isnull().sum()
#dataset.dtypes
#dataset == 0
print("Avg SAT Score of the dataset =", np.mean(dataset['SAT']))
```

```python
print(x)
```

```python
print(y)
```

##### 3. Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1)
```

##### 4. Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

# fit and predict methods both need a 2D array, so we reshape our 1D array
# x_train= x_train.reshape(-1, 1)
# x_test = x_test.reshape(-1, 1)

regressor.fit(x_train, y_train)
```

##### 5. Predicting the Test set results

```python
y_pred = regressor.predict(x_test)
```

##### 6. Visualizing the Training set results

```python
plt.scatter(x_train, y_train, color = 'orange')
plt.plot(x_train, regressor.predict(x_train), color = 'blue')

plt.title('GPA vs SAT Score (Training Set)')
plt.xlabel('SAT Score')
plt.ylabel('GPA')

plt.show()
```

##### 7. Visualizing the Test set results

```python
plt.scatter(x_train, y_train, color = 'orange')
plt.scatter(x_test, y_test, color = 'lime')
plt.scatter(x_test, y_pred, color = 'blue')
plt.plot(x_train, regressor.predict(x_train), color = 'black')

plt.title('GPA vs SAT Score (Test Set)')
plt.xlabel('SAT Score')
plt.ylabel('GPA')

plt.legend(["Training Set", "Testing Set", "Predictions for Testing Set"])

plt.show()
```

##### 8. Printing metrics for the Model

```python
from sklearn import metrics

accuracy = regressor.score(x_test, y_test)
mse = metrics.mean_squared_error(y_test, y_pred)
mae = metrics.mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
rss = ((y_pred-y_test)**2).sum()
r2_score = metrics.r2_score(y_test,y_pred)
```

```python
print("Accuracy:", accuracy)
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("Root Mean Squared Error:", rmse)
print("Residual Sum of square:", rss)
print("R-Square Error:", r2_score)
print("Coefficient:", regressor.coef_)
print("Intercept:", regressor.intercept_)
```
